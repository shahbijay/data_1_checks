# Import libraries.

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

# Import Data File

```python
housing_data = pd.read_csv('Boston_Housing_Data.csv')
housing_data
```

# Display graph using above data.

## Scatter Plot

```python
fig = plt.gcf()

# Change seaborn plot size
fig.set_size_inches(12, 6)

# Create a scatter plot using seaborn
sns.scatterplot(x='CRIM', y='MEDV', data=housing_data)

# Set plot title and labels
plt.title('Crime Rate vs Median Home Value Scatter Plot')
plt.xlabel('Crime Rate')
plt.ylabel('Median Home Value')

# Display the plot
plt.show()
```

## Heat Map

```python
# Create a heatmap using seaborn
sns.heatmap(housing_data[['CRIM', 'MEDV']].corr(), annot=True, cmap='coolwarm')

# Set plot title
plt.title('Crime Rate vs Median Home Value Heat Map')

# Display the plot
plt.show()
```

## Box Plot

```python
# Create a box plot using seaborn
sns.boxplot(data=housing_data[['CRIM', 'MEDV']], palette='Set3')

# Set plot title and labels
plt.title('Crime Rate vs Median Home Value Box Plot')
plt.xlabel('Variable')
plt.ylabel('Value')

# Display the plot
plt.show()
```

# First read all the data from URL

```python
import urllib
from urllib.request import urlretrieve
import pandas as pd

url = 'http://lib.stat.cmu.edu/datasets/boston'
urlretrieve(url, 'boston.txt')

data = []
with open('boston.txt', 'r') as file:
    for line in file:
        row = line.split()
        if len(data) == 0:
            data.append(row[:14])
        elif len(data[-1]) < 14:
            data[-1].extend(row[:14-len(data[-1])])
        else:
            data.append(row[:14])

df_web = pd.DataFrame(data, columns=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'])
print(df_web.shape)
df_web.head(25)
```

## Since the above dataframe has 9 rows that is not data, let's delete those rows and create a new dataframe called "boston_housing"

```python
# Create a new DataFrame with rows 9 to the end of table-1
boston_housing = df_web.iloc[9:, :].reset_index(drop=True)

# Print the shape of the new DataFrame
boston_housing
```
